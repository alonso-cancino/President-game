```python
import random
import math
import numpy as np
from itertools import chain, combinations
import itertools
def powerset(iterable):
    s = list(iterable)
    return list(chain.from_iterable(combinations(s, r) for r in range(len(s)+1)))
```

```python
class culo_game:
    
    def __init__(self,n_j):
        
        self.players_data = []
        self.stack = []
        self.history = []
        self.n_j = n_j
        self.deck = []
        self.active_player = None
        self.last_player = None
    
    def deal(self):
        
        available_indices =list(range(40))
        
        for pinta in ['O','B','C','E']:
            for n in range(10):
                self.deck.append(str(n)+pinta)
        
        for player in range(self.n_j):
            
            sample = random.sample(available_indices,int(40/self.n_j))
            self.players_data.append([self.deck[ind] for ind in sample])
            available_indices = list(set(available_indices) - set(sample))
            
    def active_player_0(self):
        
        if self.active_player == None:
            for player in range(self.n_j):
                if '0O' in self.players_data[player]:
                    self.active_player = player
        
    def possible_plays(self):
        
        active_hand = self.players_data[self.active_player]
        curr_quantity = len(self.stack)
        
        if curr_quantity == 0:
            if ('0O' in active_hand):
                pos_plays = [lst for lst in list(powerset([c for c in active_hand if c[0]=='0'])) if ('0O' in lst)]
                #for p in range(len(pos_plays)):
                    #print(str(p)+': ',pos_plays[p])
                return pos_plays          
            
        elif self.stack == ['*']:
            pos_plays = [lst for lst in list(powerset(active_hand)) if len(list(set([c[0] for c in lst])))==1]
            pos_plays.append([])
            return pos_plays
            
        else:
            curr_number = int(self.stack[0][0])
            pos_plays = [lst for lst in itertools.combinations(active_hand, curr_quantity) if (len(list(set([c[0] for c in lst])))==1) and (int(lst[0][0]) > curr_number)]
            
            #print([lst for lst in itertools.combinations(active_hand, curr_quantity) if (len(list(set([c[0] for c in lst])))==1)])
            pos_plays.append([])
            return pos_plays
     
    def play_select(self, selection = None):
        
        if self.last_player == self.active_player:
            self.stack = ['*']
            self.history.append('*')
            
        pos_plays = self.possible_plays()
        
        for ind in range(len(pos_plays)):
            print(ind,': ',pos_plays[ind])
        
        if selection == None:
            selection = int(input('Que jugada elige :'))
        
        if pos_plays[selection] != []:
        
            self.stack = list(pos_plays[selection])
            self.history.append(pos_plays[selection])
        
            self.last_player = self.active_player
            self.players_data[self.active_player] = list(set(self.players_data[self.active_player])-set(pos_plays[selection]))
        
        self.active_player = (self.active_player + 1)% self.n_j
        
    def hand_eval(self, player):
        
        hand_evaluation = ''
        
        for c in self.players_data[player]:
            
            hand_evaluation += -int(c[0])-1
            
        return hand_evaluation
        
```

```python
n_players = 2

hand=[[]]*n_players
game = culo_game(n_players)

game.deal()

for i in range(n_players):
    hand[i] = game.players_data[i]
    hand[i].sort()
    print('El jugador ',i,' tiene mano :\n')
    print(hand[i],'\n')

game.active_player_0()

while min(len(hand[0]),len(hand[1])) != 0 :
    print('\n')
    print('En la mesa hay:')
    print(game.stack,'\n')
    print('En tu mano tienes:')
    game.players_data[game.active_player].sort()
    print(game.players_data[game.active_player],'\n')
    print('Las jugadas posibles son:')
    pos_plays = game.possible_plays()
    print()
    input()
```

```python
###########################################################
class culo:
#############################################################
    
    def __init__(self, n_j):
        
        #######################################################
        # Genera el mazo de cartas
        #######################################################
        
        available_indices = list(range(40))
        
        self.deck = []
        for pinta in ['O','B','C','E']:
            for n in range(10):
                self.deck.append(str(n)+pinta)
        
        #######################################################
        # Genera las manos de los jugadores
        #######################################################
        
        self.n_j = n_j
        
        self.history = []
        
        self.players_data = []
        
        self.stack = []
        
        for player in range(n_j):
            
            sample = random.sample(available_indices,int(40/n_j))
            
            self.players_data.append([player,[self.deck[ind] for ind in sample]])
            
            available_indices = list(set(available_indices) - set(sample))
            
        for player in range(n_j):
            
            if '0O' in self.players_data[player][1]:
                
                self.active_player = player
                
                self.last_player = player
    
    #########################################################################################
    #
    # Parte la partida el 2 de oros con todos sus 2
    #
    #########################################################################################
    
    def start_game(self):
        
        if self.stack == []:
            
            self.stack = [c for c in self.players_data[self.active_player][1] if int(c[0])==0]
            
            self.history = [c for c in self.players_data[self.active_player][1] if int(c[0])==0] 
            
            self.players_data[self.active_player][1] = [c for c in self.players_data[self.active_player][1] if int(c[0])!=0]
            
            self.active_player = (self.active_player + 1) % self.n_j
            
        else:
            
            print('The game has already started')
            
    #########################################################################################
    
    def possible_plays(self):
        
        active_hand = self.players_data[self.active_player][1]
        
        curr_quantity = len(self.stack)
        
        if curr_quantity == 0:
            
            if ('0O' in active_hand):
                
                pos_plays = [lst for lst in list(powerset([c for c in active_hand if c[0]=='0'])) if ('0O' in lst)]
                
                for p in range(len(pos_plays)):
                    
                    print(str(p)+': ',pos_plays[p])
                    
                return pos_plays
            
            else:
                print('El jugador activo no tiene el 2 de oros.\n')
                
                self.active_player = (self.active_player + 1) % self.n_j
                
        elif active_hand == []:
            
            print('El jugador activo no tiene mano.\n')
            
        else:
            
            curr_number = int(self.stack[0][0])
            pos_plays = [lst for lst in itertools.combinations(active_hand, curr_quantity) if (len([c[0] for c in lst])==1) and (int(lst[0][0]) > curr_number)]
            pos_plays.append([])
            
            return pos_plays
        
    #########################################################################################
    
    def play_selected(self,x = None):
        
        print('Las jugadas posibles son: \n')
        
        pos_plays = self.possible_plays()
        
        #for i in range(len)
        if x== None:
            x = int(input('Elegir jugada :\n'))
            
        self.stack = list(pos_plays[x])
        
        self.players_data[self.active_player][1] = [c for c in self.players_data[self.active_player][1] if c not in list(pos_plays[x])]
        
        self.active_player = (self.active_player +1) % self.n_j
    
            
    #########################################################################################
    ##############
    # Se define la función sort_hands que ordena las manos
    ################################################
    
    #def sort_hands(self):
    #    
    #    for player in range(n_j):
    #        
    #        self.players_data[player][1]
    
            
    ################################################
    # Se define la función play(cards_played) que toma
    # una lista de cartas para jugar y las pone en el stack
    ################################################
    
    
    def play(self,cards_played=[]):
        
        curr_value = self.stack[0][0]
        curr_quantity = len(self.stack)
        active_player = self.active_player
        last_player = self.last_player
        active_hand = self.players_data[self.active_player][1]
        
        # Check de validez de la jugada
        
        cards_are_inhand = (set(cards_played)<=set(active_hand))
        lens_are_equal = (curr_quantity == len(cards_played))
        cards_are_equal = (len(list(set([c[0] for c in cards_played])))==1)
        
        if cards_played != []:
            cards_are_greater = (list(set([c[0] for c in cards_played]))[0] > curr_value)
        
        # Check de ronda en blanco
        
        ######################################################################################################################
        #     Si el jugador activo parte la ronda
        ######################################################################################################################
        
        round_starter = (active_player == last_player)
        
        #print(round_starter)
                
        if round_starter and (active_hand != []) and (cards_played != []) and cards_are_equal and cards_are_inhand:
            
            self.stack = cards_played
            
            self.players_data[active_player][1] = list(set(active_hand)-set(cards_played))
            
            for c in self.stack:
                self.history.append(c)
            
            self.last_player = active_player
            
            self.active_player = (active_player + 1) % self.n_j
            
        elif round_starter and (active_hand == []):
            
            self.last_player = (active_player + 1) % self.n_j
            
            self.active_player = (active_player + 1) % self.n_j
            
        elif round_starter and (active_hand != []) and (cards_played == []):
            
            print('Jugador ',active_player,' debe partir la ronda jugando cartas.')
        
        elif round_starter and (active_hand != []) and (cards_played != []) and (cards_are_equal == False):
            
            print('Jugador ',active_player,' intentó partir la ronda con cartas de distinto número.')
        
        elif round_starter and (active_hand != []) and (cards_played != []) and (cards_are_equal == False) and (cards_are_inhand == False):
            
            print('Jugador ',active_player,' intentó jugar cartas que no estaban en su mano.')
            
        ######################################################################################################################
        # Si el jugador activo no parte la ronda
        ###################################################################################################################### 
        
        elif ((round_starter == False) and (cards_played == [])) or ((round_starter == False) and (active_hand == [])):
            
            self.active_player = (active_player + 1) % self.n_j
           
        elif (round_starter == False) and (cards_played != []) and cards_are_inhand and lens_are_equal and cards_are_equal and cards_are_greater :
            
            self.stack = cards_played
            
            self.players_data[self.active_player][1] = list(set(active_hand)-set(cards_played))
            
            self.last_player = active_player
            
            self.active_player = (active_player + 1) % self.n_j
            
            for c in self.stack:
                
                self.history.append(c)

            
            
#Juego = culo(2)
#print(set(Juego.players_data[0][1]).intersection(set(Juego.players_data[1][1])))
#print(len(Juego.players_data[0][1]))
#print(set(Juego.players_data[1][1]).intersection(set(Juego.players_data[0][1])))
#print(len(Juego.players_data[1][1]))
```
