```python
from tkinter import *
from PIL import ImageTk,Image
import random
import numpy as np
```

```python
###########################################################
class culo:
#############################################################
    
    def __init__(self, n_j):
        
        #######################################################
        # Genera el mazo de cartas
        #######################################################
        
        available_indices = list(range(40))
        
        self.deck = []
        for pinta in ['O','B','C','E']:
            for n in range(10):
                self.deck.append(str(n)+pinta)
        
        #######################################################
        # Genera las manos de los jugadores
        #######################################################
        
        self.n_j = n_j
        
        self.players_data = []
        
        self.stack = []
        
        for player in range(n_j):
            
            sample = random.sample(available_indices,int(40/n_j))
            
            self.players_data.append([player,[self.deck[ind] for ind in sample]])
            
            available_indices = list(set(available_indices) - set(sample))
            
        for player in range(n_j):
            
            if '0O' in self.players_data[player][1]:
                
                self.active_player = player
                
                self.last_player = player
    
    ###############################################
    # Parte la partida el 2 de oros con todos sus 2
    ###############################################
    
    def start_game(self):
        
        if self.stack == []:
            
            self.stack = [c for c in self.players_data[self.active_player][1] if int(c[0])==0]
            
            self.players_data[self.active_player][1] = [c for c in self.players_data[self.active_player][1] if int(c[0])!=0]
            
            self.active_player = (self.active_player + 1) % self.n_j
            
        else:
            
            print('The game has already started')
            
    
            
    ################################################
    # Se define la función sort_hands que ordena las manos
    ################################################
    
    #def sort_hands(self):
    #    
    #    for player in range(n_j):
    #        
    #        self.players_data[player][1]
    
            
    ################################################
    # Se define la función play(cards_played) que toma
    # una lista de cartas para jugar y las pone en el stack
    ################################################
            
    def play(self,cards_played=[]):
        
        curr_value = self.stack[0][0]
        curr_quantity = len(self.stack)
        active_player = self.active_player
        last_player = self.last_player
        active_hand = self.players_data[self.active_player][1]
        
        # Check de validez de la jugada
        
        cards_are_inhand = (set(cards_played)<=set(active_hand))
        lens_are_equal = (curr_quantity == len(cards_played))
        cards_are_equal = (len(list(set([c[0] for c in cards_played])))==1)
        
        if cards_played != []:
            cards_are_greater = (list(set([c[0] for c in cards_played]))[0] > curr_value)
        
        # Check de ronda en blanco
        
        
        ######################################################################################################################
        #     Si el jugador activo parte la ronda
        ######################################################################################################################
        
        round_starter = (active_player == last_player)
        
        #print(round_starter)
                
        if round_starter and (active_hand != []) and (cards_played != []) and cards_are_equal and cards_are_inhand:
            
            self.stack = cards_played
            
            self.players_data[active_player][1] = list(set(active_hand)-set(cards_played))
            
            self.last_player = active_player
            
            self.active_player = (active_player + 1) % self.n_j
            
        elif round_starter and (active_hand == []):
            
            self.last_player = (active_player + 1) % self.n_j
            
            self.active_player = (active_player + 1) % self.n_j
            
        elif round_starter and (active_hand != []) and (cards_played == []):
            
            print('Jugador ',active_player,' debe partir la ronda jugando cartas.')
        
        elif round_starter and (active_hand != []) and (cards_played != []) and (cards_are_equal == False):
            
            print('Jugador ',active_player,' intentó partir la ronda con cartas de distinto número.')
        
        elif round_starter and (active_hand != []) and (cards_played != []) and (cards_are_equal == False) and (cards_are_inhand == False):
            
            print('Jugador ',active_player,' intentó jugar cartas que no estaban en su mano.')
            
            
        ######################################################################################################################
        # Si el jugador activo no parte la ronda
        ###################################################################################################################### 
        
        elif ((round_starter == False) and (cards_played == [])) or ((round_starter == False) and (active_hand == [])):
            
            self.active_player = (active_player + 1) % self.n_j
           
        elif (round_starter == False) and (cards_played != []) and cards_are_inhand and lens_are_equal and cards_are_equal and cards_are_greater :
            
            self.stack = cards_played
            
            self.players_data[self.active_player][1] = list(set(active_hand)-set(cards_played))
            
            self.last_player = active_player
            
            self.active_player = (active_player + 1) % self.n_j
            
```
